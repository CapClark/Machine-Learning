##Dummy Variables Creation aka One hot encoding

###Binary Values

```python
df["Exercise"] = (df["Exercise"] == "Yes").astype(int)
```

```python
df["Heart_Disease"] = (df["Heart_Disease"] == "Yes").astype(int)
```

```python
df["Skin_Cancer"] = (df["Skin_Cancer"] == "Yes").astype(int)
```

```python
df["Other_Cancer"] = (df["Other_Cancer"] == "Yes").astype(int)
```

```python
df["Depression"] = (df["Depression"] == "Yes").astype(int)
```

```python
df["Diabetes"] = (df["Diabetes"] == "Yes").astype(int)
```

```python
df["Arthritis"] = (df["Arthritis"] == "Yes").astype(int)
```

```python
df["Smoking_History"] = (df["Smoking_History"] == "Yes").astype(int)
```

```python
df["Sex"] = (df["Sex"] == "Female").astype(int)
```

```python
df
```

###Dummy Variable for Checkup



```python
checkUpDum = pd.get_dummies(df.Checkup)
#Merging the dummy with the exisiting graph
mergedDF = pd.concat([df, checkUpDum], axis = "columns")
df.drop(columns = 'Checkup')
mergedDF
```

```python
#Dropping one column to avoid multicollinearity
mergedDF = mergedDF.drop(columns = 'Within the past 2 years')
```

###Dummy Variable for General Health

```python
GenHealthDum = pd.get_dummies(df.General_Health)
mergedDF = pd.concat([df, GenHealthDum], axis = 'columns')
mergedDF = mergedDF.drop(columns = "Very Good")
mergedDF = mergedDF.drop(columns = "Checkup")
mergedDF
```

```python
# Create a LabelEncoder instance
label_encoder = LabelEncoder()
# Apply label encoding to the "General_Health" column
mergedDF["General_Health"] = label_encoder.fit_transform(mergedDF["General_Health"])
mergedDF["Age_Category"] = label_encoder.fit_transform(mergedDF["Age_Category"])
mergedDF
```

##Training Model

```python
model = DecisionTreeClassifier()
X = mergedDF.drop(columns = 'General_Health')
y = mergedDF['General_Health']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model.fit(X_train,y_train)
```

```python

# Your input data as a 1D array
input_data = np.array([1, 0, 0, 0, 0, 1, 0, 0, 2, 157, 61.23, 24.69, 1, 4.0, 40.0, 8.0, 4.0, 0, 0,0,0])

# Reshape the input data into a 2D array
reshaped_data = input_data.reshape(1, -1)
predictions = model.predict(reshaped_data)
predictions
```

##Measuring Model's Accuracy


```python
predictions = model.predict(X_test)

score = accuracy_score(y_test, predictions)
score
```
